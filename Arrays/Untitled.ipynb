## 1. Reverse an Array


```python
def reverse(arr):
    n=len(arr)
    i=0
    j=n-1
    while i<j:
        arr[i],arr[j]=arr[j],arr[i]
        i+=1
        j-=1

    return arr


arr=[1,2,3,4,5,6]
reverse(arr)
```

## 2. Find the maximum and minimum element


```python
def max_min(arr):
    minm=arr[0]
    maxm=arr[0]
    for i in range(len(arr)):
        if arr[i]>maxm:
            maxm=arr[i]
        elif arr[i]<minm:
            minm=arr[i]

    return maxm,minm



arr=[1,2,3,4,5,6]
maxm,minm = max_min(arr)
print("maxm:",maxm)
print("minm: ",minm)
    
```

## 3. sum of elements

```python
def sum_of_elements(arr):
    sum=0
    for i in arr:
        sum+=i

    return sum  



arr=[1,2,3,4,5]
sum_of_elements(arr)
    
```

## 4. Second largest

```python
def sec_lar(arr):
    largest=float("-inf")
    sec=float("-inf")
    for num in arr:
        if num>largest:
            sec=largest
            largest=num
        elif num>sec and num!=largest:
            sec=num
    return sec



arr=[1,2,3,4,5]
sec_lar(arr)
```

## 5. frequency of elements


```python
def count_freq(arr):
    freq={}
    for num in arr:
        if num in freq:
            freq[num]+=1
        else:
            freq[num]=1
    return freq        

arr=[1,1,2,2,2,3,45,4,3,4]
count_freq(arr)


        
```

## 6. check if array is sorted

```python
def check(arr):
    for i in range(len(arr)-1):
        if arr[i]>arr[i+1]:
            return False
    return True


arr=[1,2,4,6,8]
if check(arr):
    print("sorted")
else:
    print("Not Sorted")

    

            
```

## 7. Rotate the array to the right by k positions


```python
def rot_right(arr,k):
    n=len(arr)
    temp=arr[-k:]
    for i in range(n-k-1,-1,-1):
        arr[i+k]=arr[i]

    for i in range(k):
        arr[i]=temp[i]


    return arr



arr=[1,2,3,4,5]
k=2
rot_right(arr,k)
    


            
        
        

    
```

## 8. Find pair with given sum

```python
def pair(arr,target):
    for i in range(len(arr)):
        for j in range(i+1):
            if arr[i]+arr[j]==target:
                return (arr[i],arr[j])


arr=[1,2,3,4,5]
target=3
pair(arr,target)
        
```

## 9. Remove Duplicates 

```python
def remove_dup(arr):
    res=[]
    for num in arr:
        if num not in res:
            res.append(num)
    return res


arr=[1,1,3,4,5,3,4]
remove_dup(arr)
```

## 10. Find missing Number

```python
def find(arr):
    n=len(arr)+1
    total_sum=n*(n+1)//2
    act_sum=0
    for num in arr:
        act_sum+=num

    return total_sum-act_sum


arr=[1,2,4,5]
find(arr)
    
```
